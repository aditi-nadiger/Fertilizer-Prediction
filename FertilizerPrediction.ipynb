```python
#Defining function for Continuous and catogorical variable
def plot_conti(x):
    fig, axes = plt.subplots(nrows=1,ncols=3,figsize=(15,5),tight_layout=True)
    axes[0].set_title('Histogram')
    sns.histplot(x,ax=axes[0])
    axes[1].set_title('Checking Outliers')
    sns.boxplot(x,ax=axes[1])
    axes[2].set_title('Relation with output variable')
    sns.boxplot(y = x,x = df['Fertilizer Name'])
    
def plot_cato(x):
    fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(15,5),tight_layout=True)
    axes[0].set_title('Count Plot')
    sns.countplot(x,ax=axes[0])
    axes[1].set_title('Relation with output variable')
    sns.countplot(x = x,hue = df['Fertilizer Name'], ax=axes[1])
```

```python
#outliers
for col in cols:
    # Access the column data
    column = df[col]
    # Calculate the quartiles
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    # Calculate the IQR (Interquartile Range)
    IQR = Q3 - Q1
    
    # Calculate the lower and upper bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify the outliers in the column
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    
    # Calculate the number of outliers
    num_outliers = len(outliers)
    
    # Print the number of outliers in the current column
    print(f"Column '{col}' has {num_outliers} outliers.")
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Assuming X and y are your features and target variable
print('Original dataset shape %s' % Counter(y))

xts, yts = smote.fit_resample(x, y)

print('Resampled dataset shape %s' % Counter(yts))
print('Original shape:', x.shape)
print('Resampled shape:', xts.shape)
```

```python
#STANDARDISATION
# diff bte standardisation and normalisation
```

```python
#importing libraries for data transformation
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xts = scaler.fit_transform(xts)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(xts, yts, test_size=0.2, random_state=42)
```

```python
print("Samples in Training data ", len(X_train))
print("Samples in Testing data ", len(X_test))
```

```python
#importing models and evaluation metrics
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, auc
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# Assuming X_train, X_test, y_train, y_test are already defined
gbc = GradientBoostingClassifier(random_state=42)
gbc.fit(X_train, y_train)
y_pred_gbc = gbc.predict(X_test)
y_train_pred_gbc = gbc.predict(X_train)

# Getting all scores for Gradient Boosting Classifier
train_accuracy_gbc = round(accuracy_score(y_train, y_train_pred_gbc), 3)
accuracy_gbc = round(accuracy_score(y_test, y_pred_gbc), 3)
precision_score_gbc = round(precision_score(y_test, y_pred_gbc, average='weighted'), 3)
recall_score_gbc = round(recall_score(y_test, y_pred_gbc, average='weighted'), 3)
f1_score_gbc = round(f1_score(y_test, y_pred_gbc, average='weighted'), 3)
roc_score_gbc = round(roc_auc_score(y_test, gbc.predict_proba(X_test), multi_class='ovr'), 3)

print("The accuracy on train data is ", train_accuracy_gbc)
print("The accuracy on test data is ", accuracy_gbc)
print("The precision on test data is ", precision_score_gbc)
print("The recall on test data is ", recall_score_gbc)
print("The f1 on test data is ", f1_score_gbc)
print("The roc_score on test data is ", roc_score_gbc)

cm_gbc = confusion_matrix(y_test, y_pred_gbc)
print(cm_gbc)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# Train the model
xgb = XGBClassifier()
xgb.fit(X_train, y_train)

# Make predictions
y_pred_xgb = xgb.predict(X_test)
y_train_pred_xgb = xgb.predict(X_train)

# Calculate scores
train_accuracy_xgb = round(accuracy_score(y_train, y_train_pred_xgb), 3)
accuracy_xgb = round(accuracy_score(y_test, y_pred_xgb), 3)
precision_score_xgb = round(precision_score(y_test, y_pred_xgb, average='weighted'), 3)
recall_score_xgb = round(recall_score(y_test, y_pred_xgb, average='weighted'), 3)
f1_score_xgb = round(f1_score(y_test, y_pred_xgb, average='weighted'), 3)

# For multi-class classification, roc_auc_score requires probability estimates
y_pred_proba_xgb = xgb.predict_proba(X_test)
roc_score_xgb = round(roc_auc_score(y_test, y_pred_proba_xgb, multi_class='ovr'), 3)

# Print scores
print("The accuracy on train data is ", train_accuracy_xgb)
print("The accuracy on test data is ", accuracy_xgb)
print("The precision on test data is ", precision_score_xgb)
print("The recall on test data is ", recall_score_xgb)
print("The f1 on test data is ", f1_score_xgb)
print("The roc_score on test data is ", roc_score_xgb)

# Confusion matrix
cm_xgb = confusion_matrix(y_test, y_pred_xgb)
print(cm_xgb)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators= 100, criterion = 'gini' , random_state= 42)  # Example: You can adjust parameters here if needed
rfc.fit(X_train, y_train)
y_pred_rfc = rfc.predict(X_test)
# Assuming rfc is your trained Random Forest Classifier
rfc.fit(X_train, y_train)

# Predictions
y_train_pred_rfc = rfc.predict(X_train)
y_pred_rfc = rfc.predict(X_test)

# Accuracy
train_accuracy_rfc = round(accuracy_score(y_train, y_train_pred_rfc), 3)
accuracy_rfc = round(accuracy_score(y_test, y_pred_rfc), 3)

# Precision, Recall, F1-score (for each class)
precision_score_rfc = round(precision_score(y_test, y_pred_rfc, average='weighted'), 3)
recall_score_rfc = round(recall_score(y_test, y_pred_rfc, average='weighted'), 3)
f1_score_rfc = round(f1_score(y_test, y_pred_rfc, average='weighted'), 3)

# ROC AUC Score
# For multiclass ROC AUC, sklearn expects one-hot encoded or probability predictions
# Assuming y_pred_proba_rfc contains the predicted probabilities for each class
y_pred_proba_rfc = rfc.predict_proba(X_test)
roc_score_rfc = round(roc_auc_score(y_test, y_pred_proba_rfc, multi_class='ovr'), 3)

# Confusion Matrix
cm_rfc = confusion_matrix(y_test, y_pred_rfc)
print("Confusion Matrix:")
print(cm_rfc)

# Print all scores
print("The accuracy on train data is ", train_accuracy_rfc)
print("The accuracy on test data is ", accuracy_rfc)
print("The precision on test data is ", precision_score_rfc)
print("The recall on test data is ", recall_score_rfc)
print("The f1-score on test data is ", f1_score_rfc)
print("The ROC AUC score on test data is ", roc_score_rfc)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix

# Train the model
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)

# Make predictions
y_pred_dtc = dtc.predict(X_test)
y_train_pred_dtc = dtc.predict(X_train)

# Calculate scores
train_accuracy_dtc = round(accuracy_score(y_train, y_train_pred_dtc), 3)
accuracy_dtc = round(accuracy_score(y_test, y_pred_dtc), 3)
precision_score_dtc = round(precision_score(y_test, y_pred_dtc, average='weighted'), 3)
recall_score_dtc = round(recall_score(y_test, y_pred_dtc, average='weighted'), 3)
f1_score_dtc = round(f1_score(y_test, y_pred_dtc, average='weighted'), 3)

# For multi-class classification, roc_auc_score requires probability estimates
y_pred_proba_dtc = dtc.predict_proba(X_test)
roc_score_dtc = round(roc_auc_score(y_test, y_pred_proba_dtc, multi_class='ovr'), 3)

# Print scores
print("The accuracy on train data is ", train_accuracy_dtc)
print("The accuracy on test data is ", accuracy_dtc)
print("The precision on test data is ", precision_score_dtc)
print("The recall on test data is ", recall_score_dtc)
print("The f1 on test data is ", f1_score_dtc)
print("The roc_score on test data is ", roc_score_dtc)

# Confusion matrix
cm_dt = confusion_matrix(y_test, y_pred_dtc)
print(cm_dt)
```
